### 2a.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gamma = 1
omega0 = 10
omega = np.linspace(0,20,100)
N = np.sqrt((gamma*gamma/(omega0*omega0 - omega*1.0j - omega*omega)))
```

```python
plt.plot(np.real(omega), np.real(N), label='$n$')
plt.plot(np.real(omega), np.imag(N), label='$\kappa$')
plt.xlabel("omegap")
plt.ylabel("$\kappa, n$")
plt.legend()

plt.show()
```

### 2b.

```python
omega = np.linspace(0, 30, 100)
n = np.sqrt((0.5 / (100 - omega*1.0j - omega**2)) + (0.25 / (15**2 - omega*1.0j - omega**2)) + (0.25 / (25**2 - 3*omega*1.0j - omega**2)))

plt.plot(np.real(omega), np.real(n), label='Real Part')
plt.plot(np.real(omega), np.imag(n), label='Imaginary Part')

plt.xlabel('$\omega$')
plt.ylabel('n')
plt.title('Real and Imaginary Parts of n')
plt.legend()

plt.show()
```
