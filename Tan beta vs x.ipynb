```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4*np.pi, 4*np.pi, 1000)


y_tan = np.tan(x)
y_x = x

# Plot the functions
plt.plot(x, y_tan, label='tan(x)', color='blue')
plt.plot(x, y_x, label='x', color='red')
plt.ylim(-15,15)
# Find the intersection points (where tan(x) = x)
intersection_points = [(x_val, y_tan[index]) for index, x_val in enumerate(x) if abs(y_tan[index] - y_x[index]) < 0.08]

# Plot the intersection points
for point in intersection_points:
    plt.plot(point[0], point[1], 'go')

# Set the plot title and labels
plt.title('Plot of tan(x) and x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
